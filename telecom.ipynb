```
import pandas as pd
from scipy import stats

# 1. Load the dataset
df = pd.read_csv('telecom-data-set.csv')

# 2. Create a Contingency Table for Payment Method vs Customer Churn
contingency_table = pd.crosstab(df['PaymentMethod'], df['Churn'])
print("--- Contingency Table ---")
print(contingency_table)
print("\n-------------------------\n")

# 3. Perform the Chi-Square Test of Independence
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")

# 4. Interpret the results based on Alpha level (0.05)
if p < 0.05:
    print("\nResult: There is a statistically significant relationship between Payment Method and Customer Churn! (We reject the Null Hypothesis)")
else:
    print("\nResult: There is no significant relationship between Payment Method and Customer Churn.")
```
